```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Get the file path for running team one
team_one_file_path = os.path.join('../','Resources', 'team-1.csv')
```

```python
# Read the CSV file into a DataFrame
team_one_df = pd.read_csv(team_one_file_path)
team_one_df
```

```python
# Melt the DataFrame
team_one_df.melt()
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team_one_df.melt(id_vars="runner_name")
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team_one_df.melt(id_vars="coach_name")
```

```python
# Renaming the columns
team_one_df.melt(id_vars="runner_name", var_name="run", value_name="length")
```

```python
# Remove coach name from DataFrame
del team_one_df['coach_name']
team_one_df
```

```python
# Show final result
output_df = team_one_df.melt(id_vars="runner_name", var_name="run", value_name="length")
```

```python
# Save results to a CSV file
output_df.to_csv("melted_running_df.csv", sep=',')
```
